```python
from PIL import Image as pili, ImageOps as piliops

imageLocation = "images/" #image directory
borderLocation = "images/borders/" #folder location of border images

dialogueLocation = "images/dialogues/" #dialogue location for english
hindiDialogues = 'images/hindiDialogues/' #dialogue location for hindi

randomHorizontal = "images/randomHorizontal/" #random Horizontal collage location for english
randomVertical = "images/randomVertical/" #random Vertical collage location for english

randomHindiHorizontal = "images/randomHindiHorizontal/" #random Horizontal collage location for hindi
randomHindiVertical = "images/randomHindiVertical/" #random Vertical collage location for hindi

imageName = [0] #naming all images, first value giver for easy understanding

for i in range(1,74): #running loop for number of images
    imageName.append("image_" + str(i)+ ".jpg")
print(imageName[1]) #verifying image name
print(len(imageName)-1) #verifying number of image names
```

Creating ENGLISH Dialogues on images


```python
from PIL import Image as pili, ImageDraw as pild, ImageFont as pilf 

TINT_COLOR = (0, 0, 0)  # Black color on subtitle rectangle

TRANSPARENCY = .25  # Degree of transparency, 0-100%
OPACITY = int(255 * TRANSPARENCY)

for i in range(1,74): #loop running for number of images and creating subtitle for all of them
    from PIL import Image, ImageFont, ImageDraw
    img = pili.open(borderLocation + imageName[i]).convert('RGBA')
    overlay = pili.new('RGBA', img.size, (TINT_COLOR+(0,)))
    draw = pild.Draw(overlay)
    font = pilf.truetype("arial.ttf", 35) #for english font 
    text = data.values[i][0]
    w, h = font.getsize(text)
    num_lines = len(text.split('\n')) #splitting subtitles
    print(num_lines, w, h, i) #printing no.of subtitle lines, width, height, no.of image
    x, y = 25, img.height - (num_lines-0.1*num_lines)*h
    draw.rectangle((x, y-15, x + img.width - 50, y + (num_lines-0.1*num_lines)*h), fill=TINT_COLOR+(OPACITY,)) #drawing subtitle box
    draw.text((x+35, y-15), text, fill=(209, 239, 8), font=font) #writing texts
    img = pili.alpha_composite(img, overlay)
    img = img.convert("RGB") # Remove alpha for saving in jpg format.
    img.save(dialogueLocation + imageName[i]) #saving english image
```

Creating random english Horizontal and vertical stripes

```python
import numpy as np
import PIL
from os import listdir
from os.path import isfile, join

mypath = dialogueLocation
#onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
onlyfiles = [] #image Locations
verticalImage = [] #to run loop for creating vertical stack
imageCount = [] #image paths for vetical stack, to be covnerted to pdf
for i in range(1,74): 
    onlyfiles.append(engList[i])
i, j = 1, 1
while len(onlyfiles) > 0: #running loop for vertical stack of horizontal stack created
    lines = 0
    print("new page!")
    num_cols = old_num_cols = 0
    print(len(onlyfiles))
    
    while lines < 6 and 0 < len(onlyfiles): #running loop for horizontal stack
        lines += 1
        while num_cols == old_num_cols:
            num_cols = np.random.randint(3, 6)
        old_num_cols = num_cols
        list_im = onlyfiles[:num_cols]
        imgs = [ PIL.Image.open(k) for k in list_im ]
        
        # pick the image which is the smallest, and resize the others to match it (can be arbitrary image shape here)
        min_shape = sorted( [(np.sum(k.size), k.size ) for k in imgs])[0][1]
        imgs_comb = np.hstack( (np.asarray( k.resize(min_shape) ) for k in imgs ) )
        
        # save that beautiful picture
        imgs_comb = PIL.Image.fromarray( imgs_comb)
        imgs_comb.save( randomHorizontal + imageName[i] ) 
        
        verticalImage.append(randomHorizontal + imageName[i])
        
        #print(window)   
        onlyfiles = onlyfiles[num_cols:]    
        i = i+1
        
    imgs = [ PIL.Image.open(i) for i in verticalImage ]

    # pick the image which is the largest, and resize the others to match it 
    min_shape = sorted( [(np.sum(k.size), k.size ) for k in imgs], reverse = 1)[0][1]
    imgs_comb = np.vstack( (np.asarray( k.resize(min_shape) ) for k in imgs ) )

    # save that beautiful picture
    imgs_comb = PIL.Image.fromarray( imgs_comb)
    imgs_comb.save( randomVertical + imageName[j], width=650, height=1300 )
    imageCount.append(randomVertical + imageName[j])
    j += 1
    verticalImage = verticalImage[len(verticalImage):]
    
print(imageCount)
```

converting random english images to pdf

```python
from PIL import Image
from fpdf import FPDF

#pdf = FPDF()
cover = Image.open('images/cover.jpg')
width, height = cover.size
pdf = FPDF(unit = "pt", format = [width, height])
imagelist = ['images/cover.jpg']
    
# imagelist is the list with all image filenames
for x in range(1, len(imageCount)+1):
    imagelist.append(randomVertical + imageName[x])
    

for image in imagelist:
    pdf.add_page()
    pdf.image(image, 0, 0, width, height)
pdf.output("result.pdf", "F")
print("done!")
```
